```
"""concat sub folder"""
import pandas as pd
import glob as glob

list_class=['adult','horror','violence']
for obj in list_class:
  all_filenames=[i for i in glob.glob('/content/gdrive/My Drive/YoutubeContentAnalysis/'+obj+'/*.csv')]
  combined_csv=pd.concat([pd.read_csv(f) for f in all_filenames])

  combined_csv.to_csv("/content/gdrive/My Drive/YoutubeContentAnalysis/"+obj+".csv",index=False,encoding='utf-8-sig')
```
